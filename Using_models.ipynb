# BERT-BankingClassifier

```python
from transformers import pipeline, AutoTokenizer

text = "i forgot PIN"
model_id = "saribasmetehan/BERT-BankingClassifier"

classifier = pipeline("text-classification", model=model_id)

preds = classifier(text)
print(preds)
```

```python
text_1 = "My money transfer was rejected"
preds = classifier(text_1)
print(preds)
```

# Spacy-BankingClassifier

```python
import spacy

model_path = "/content/drive/MyDrive/spaCy_textcat/output/model-best"

nlp = spacy.load(model_path)

doc = nlp(text)

max_label = max(doc.cats, key=doc.cats.get)
max_score = doc.cats[max_label]

result = [{'label': max_label, 'score': max_score}]
print(result)
```
